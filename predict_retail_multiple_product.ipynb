```python
df = pd.read_csv('online_retail_II.csv', delimiter = ";")
df
```

```python
df = df[df['Country'] == 'United Kingdom']
df['InvoiceDate'] = df['InvoiceDate'].str.slice(0, 10)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df = df.groupby(['StockCode', 'InvoiceDate']).Quantity.sum().reset_index()
df
```

```python
df1 = df.groupby('StockCode').resample('W-Mon', on='InvoiceDate').sum().reset_index().sort_values(by='InvoiceDate')
df1
```

```python
df2 = df1.pivot(index='InvoiceDate', columns='StockCode', values='Quantity')
df2
```

```python
df3 = df1[df1['StockCode'] == "22069"]
df3.drop('StockCode', axis = 1)
df3
```

```python
m = Prophet()
m.fit(df3)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
